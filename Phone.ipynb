#Aggregate

```python

path1 = r"C:\\Users\\hariv\\OneDrive\\Desktop\\Data Science\\Python\\Phone_Pe\\pulse\\data\\aggregated\\transaction\\country\\india\\state"
Agg_state1 = os.listdir(path1)

columns1 = {'State': [], 'Year': [], 'Quarter': [], 'Transaction_type': [], 'Transaction_count': [],
            'Transaction_amount': []}

for state in Agg_state1:
    cur_state = os.path.join(path1, state)  # Correct path construction using os.path.join()
    agg_year_list = os.listdir(cur_state)
    
    for year in agg_year_list:
        cur_year = os.path.join(cur_state, year)  # Correct path construction using os.path.join()
        agg_file_list = os.listdir(cur_year)
        
        for file in agg_file_list:
            cur_file = os.path.join(cur_year, file)  # Correct path construction using os.path.join()
            with open(cur_file, 'r') as data:
                A = json.load(data)
                for i in A['data']['transactionData']:
                    name = i['name']
                    for pay in i["paymentInstruments"]:
                        count = pay['count']
                        amount = pay['amount']
                        columns1['Transaction_type'].append(name)
                        columns1['Transaction_count'].append(count)
                        columns1['Transaction_amount'].append(amount)
                        columns1['State'].append(state)
                        columns1['Year'].append(year)
                        columns1['Quarter'].append(int(os.path.splitext(file)[0]))
                
df_agg_trans = pd.DataFrame(columns1)
```

```python
path2 = r"C:\\Users\\hariv\\OneDrive\\Desktop\\Data Science\\Python\\Phone_Pe\\pulse\\data\\aggregated\\user\\country\\india\\state"
Agg_state2 = os.listdir(path2)

columns2 = {'State': [], 'Year': [], 'Quarter': [], 'Brand name': [], 'Count': [],
            'Percentage': []}

for state in Agg_state2:
    cur_state = os.path.join(path2, state)  # Correct path construction using os.path.join()
    agg_year_list = os.listdir(cur_state)
    
    for year in agg_year_list:
        cur_year = os.path.join(cur_state, year)  # Correct path construction using os.path.join()
        agg_file_list = os.listdir(cur_year)

        for file in agg_file_list:
            cur_file = os.path.join(cur_year, file)  # Correct path construction using os.path.join()
            try:
                with open(cur_file, 'r') as data:
                    B = json.load(data)
                    for i in B["data"]["usersByDevice"]:
                        brand_name = i["brand"]  
                        counts = i["count"]  
                        percents = i["percentage"]  
                        columns2["Brand name"].append(brand_name)
                        columns2["Count"].append(counts)
                        columns2["Percentage"].append(percents)
                        columns2["State"].append(state)
                        columns2["Year"].append(year)
                        columns2["Quarter"].append(int(os.path.splitext(file)[0]))
            except:
                pass

df_agg_user = pd.DataFrame(columns2)
```

```python

path3 = r"C:\Users\hariv\OneDrive\Desktop\Data Science\Python\Phone_Pe\pulse\data\aggregated\insurance\country\india\state"
Agg_state3 = os.listdir(path3)
columns3 = {'State': [], 'Year': [], 'Quarter': [], 'Count': [],
            'Amount': [],'Name':[]}
for state in Agg_state3:
    cur_state = os.path.join(path3, state)
    map_year_list = os.listdir(cur_state)
    for year in map_year_list:
        cur_year = os.path.join(cur_state, year)
        map_file_list = os.listdir(cur_year)
        for file in map_file_list:
            cur_file = os.path.join(cur_year, file)
            with open(cur_file, "r") as data:
                C = json.load(data)
                # Extracting data for insurance transactions
                for transaction in C['data']['transactionData']:
                    if transaction['name'] == 'Insurance':
                        for payment_instrument in transaction['paymentInstruments']:
                            columns3['State'].append(state)
                            columns3['Year'].append(year)
                            columns3['Quarter'].append(int(os.path.splitext(file)[0]))
                            columns3['Name'].append(transaction['name'])
                            columns3['Count'].append(payment_instrument['count'])
                            columns3['Amount'].append(payment_instrument['amount'])

df_agg_insu= pd.DataFrame(columns3)
                    
```

#map

```python
path4 = r"C:\\Users\\hariv\\OneDrive\\Desktop\\Data Science\\Python\\Phone_Pe\\pulse\\data\\map\\transaction\\hover\\country\\india\\state"
Map_state1 = os.listdir(path4)

columns4 = {'State': [], 'Year': [], 'Quarter': [], 'District': [], 'Count': [],
            'Amount': []}

for state in Map_state1:
    cur_state = os.path.join(path4, state)
    map_year_list = os.listdir(cur_state)
    for year in map_year_list:
        cur_year = os.path.join(cur_state, year)
        map_file_list = os.listdir(cur_year)
        for file in map_file_list:
            cur_file = os.path.join(cur_year, file)
            with open(cur_file, "r") as data:
                D = json.load(data)
                for i in D['data']["hoverDataList"]:
                    name = i['name']
                    for j in i["metric"]:
                        count = j["count"]
                        amount = j["amount"]
                        columns4["State"].append(state)
                        columns4["Year"].append(year)
                        columns4["District"].append(name)  # Fix the typo here
                        columns4["Count"].append(count)
                        columns4["Amount"].append(amount)
                        columns4['Quarter'].append(int(os.path.splitext(file)[0]))
df_map_trans = pd.DataFrame(columns4)

```

```python
path5 = r"C:\\Users\\hariv\\OneDrive\\Desktop\\Data Science\\Python\\Phone_Pe\\pulse\\data\\map\\user\\hover\\country\\india\\state"
Map_state2 = os.listdir(path5)

columns5={'State':[],'Year':[],'Quarter':[],'District':[],'Registerd user':[],'Appopens':[]}

for state in Map_state2:
    cur_state=os.path.join(path5,state)
    map_year_list=os.listdir(cur_state)
    for year in map_year_list:
        cur_year=os.path.join(cur_state,year)
        map_file_list=os.listdir(cur_year)
        for file in map_file_list:
            cur_file=os.path.join(cur_year,file)
            with open(cur_file,'r')as data:
                E=json.load(data)
                for district,i in E['data']['hoverData'].items():
                    registred=i['registeredUsers']
                    appopen=i['appOpens']
                    columns5["District"].append(district)
                    columns5["Registerd user"].append(registred)
                    columns5["Appopens"].append(appopen)
                    columns5["State"].append(state)
                    columns5["Year"].append(year)
                    columns5["Quarter"].append(int(os.path.splitext(file)[0]))

df_map_user=pd.DataFrame(columns5)
```
